```python
import pandas as pd
import json
import sqlite3
import csv
import numpy as np
from shapely.geometry import shape, Point
```

```python
f = '/Volumes/USB_CHA/Hackathon/data/UTSEUS-MOBIKE-shanghai_full.csv'
```

```python
mobike = pd.read_csv(f, sep=',')
```

```python
data = mobike.sample(500) #remove sample for final data
data.head()
```

```python
with open('/Users/charlotte/Documents/UTSEUS/Cours/UM02/Projet/district_connection/shanghai_district.geojson') as f:
    js = json.load(f)
```

```python
def search_district_name(longitude, latitude, json_file):
    point = Point(longitude, latitude)
    for feature in json_file['features']:
        polygon = shape(feature['geometry'])
        if polygon.contains(point):
            return feature['properties']['NAME']
    return -1
```

```python
data['district_start'] = np.vectorize(search_district_name)(data['start_location_x'], data['start_location_y'], js)
```

```python
data['district_end'] = np.vectorize(search_district_name)(data['end_location_x'], data['end_location_y'], js)
```

```python
data_district = data[['district_start','district_end']]
```

```python
data_district_clean = data_district[data['district_start'] != data['district_end']]
```

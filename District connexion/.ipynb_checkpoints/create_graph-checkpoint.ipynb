```python
G=nx.Graph()
```

```python
with open('/Users/charlotte/Documents/UTSEUS/Cours/UM02/Projet/district_connection/district_connection.csv','r') as fp:
    lines = fp.read().split('\n')
```

```python
for line in lines:
    data = line.split(',')
    start = data[1]
    end = data[2]
    if((start != '-1') & (end != '-1')):
        G.add_node(start)
        G.add_node(end)
        G.add_edge(start, end)
        print start, end
nx.write_gexf(G,'district_connexion.gexf')
```

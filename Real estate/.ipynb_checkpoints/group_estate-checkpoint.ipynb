```python
data = estate.sample(500) #remove sample for final data
data.head()
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
data_surf = df.loc[(df['surface'] >= 25) & (df['surface'] <= 325)]
```

```python
data_surf.describe()
```

```python
data_surf_price = data_surf.loc[(data_surf['price'] >= 500000) & (data_surf['price'] <= 35000000)]
```

```python
data_surf_price.describe()
```

```python
class_surf = {}
class_surf ["<30"] = data_surf_price.loc[(data_surf_price['surface'] <= 30)]
```

```python
i = 30
while i < 300:
    class_surf[str(i)] = data_surf_price.loc[(data_surf_price['surface'] > i) & (data_surf_price['surface'] <= i+30)]
    i = i +30
```

```python
class_surf[">300"]=data_surf_price.loc[(data_surf_price['surface'] > 300)]
```

```python
for key, value in class_surf.iteritems():
    print key, value["price"].mean()
```
